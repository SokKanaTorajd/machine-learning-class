## Feature Engineering
Pengolahan Fitur

```python
from sklearn.datasets import load_iris
data = load_iris()
feature = data['data']
```

```python
data.keys()
```

```python
import pandas as pd
dfiris = pd.DataFrame(feature ,columns=['sepallength','sepalwidth',
'petallength','petalwidth'])
```

```python
dfiris.info()
```

```python
sa=0.5*dfiris['sepallength']*dfiris['sepalwidth']
pa=0.5*dfiris['petallength']*dfiris['petalwidth']
sp=2*(dfiris['sepallength']+dfiris['sepalwidth'])
pp=2*(dfiris['petallength']+dfiris['petalwidth'])
```

```python
dfiris = dfiris.assign(sepalarea=sa)
dfiris = dfiris.assign(petalarea=pa)
dfiris = dfiris.assign(sepalperimeter=sp)
dfiris = dfiris.assign(petalperimeter=pp)
```

```python
dfiris
```

```python
from scipy.stats import pearsonr
```

```python
target = data['target']
```

```python
target
```

```python
print(pearsonr(dfiris['petalarea'], target))
print(pearsonr(dfiris['sepalarea'], target))
print(pearsonr(dfiris['petalperimeter'], target))
print(pearsonr(dfiris['sepalperimeter'], target))
```

```python
print(pearsonr(dfiris['petallength'], target))
print(pearsonr(dfiris['sepallength'], target))
print(pearsonr(dfiris['petalwidth'], target))
print(pearsonr(dfiris['sepalwidth'], target))
```
